```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete
```

## School Summary

```python
#calculate average math scores
avg_math_score = school_data_complete["math_score"].mean()

avg_math_score
```

```python
#calculate average reading score 
avg_reading_score = school_data_complete["reading_score"].mean()

avg_reading_score
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_passing_rate = (avg_math_score + avg_reading_score)/2

overall_passing_rate
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
school_data_complete["passing_math"] = school_data_complete["math_score"] >= 70
school_data_complete["passing_reading"] = school_data_complete["reading_score"] >=70

percent_pass_math = ((school_data_complete["passing_math"]).mean())*100

percent_pass_math
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
percent_pass_reading = ((school_data_complete["passing_reading"]).mean())*100

percent_pass_reading
```

```python
##Create a dataframe to hold the above results
#Optional: give the displayed data cleaner formatting
final_results = [{"Total_Schools": total_num_schools, 
            "Total_Students": total_num_students, 
            "Total_Budget": total_budget, 
            "Average_Math_Score":  round(avg_math_score,2), 
            "Average_Reading_Score":  round(avg_reading_score,2), 
           "%Passing_Math": round(percent_pass_math,2),
           "%Passing_Reading": round(percent_pass_reading,2),
           "Overall_Passing_Rate": round(overall_passing_rate,2)}]

final_summary_table = pd.DataFrame(final_results)

#Arrange columns
final_summary_table = final_summary_table[[
                                     "Total_Schools",
                                     "Total_Students",
                                     "Total_Budget",
                                     "Average_Math_Score",
                                     "Average_Reading_Score",
                                     "%Passing_Math",
                                     "%Passing_Reading",
                                     "Overall_Passing_Rate" ]]



#Display
final_summary_table

```

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#define school_data_completE
school_data_complete["passing_math"] = school_data_complete["math_score"] >= 70
school_data_complete["passing_reading"] = school_data_complete["reading_score"] >= 70

school_data_complete
```

```python
#group school name by using the groupby function
grouped_school_summary = school_data_complete.groupby(["school_name"]).mean()
grouped_school_summary["Per Student Budget"] = round(grouped_school_summary["budget"]/grouped_school_summary["size"],2)
grouped_school_summary["% Passing Math"] = round(grouped_school_summary["passing_math"]*100,2)
grouped_school_summary["% Passing Reading"] = round(grouped_school_summary["passing_reading"]*100,2)
grouped_school_summary["Average Math Score"] = round(grouped_school_summary["math_score"].mean(),2)
grouped_school_summary["Average Reading Score"] = round(grouped_school_summary["reading_score"].mean(),2)
grouped_school_summary["% Overall Passing Rate"] = round(((grouped_school_summary["passing_math"] + 
                                                       grouped_school_summary["passing_reading"])/2)*100,2)

#merge grouped_school_summary with school_data
school_summary_data = pd.merge(grouped_school_summary, school_data, how="left", on=["school_name" , "school_name"])


#create a dataframe to store result
df_school_summary_data = pd.DataFrame({"School Name":school_summary_data["school_name"],
                                                "School Type":school_summary_data["type"],
                                                "Total Students":school_summary_data["size_y"],
                                                "Total School Budget":school_summary_data["budget_x"],
                                                "Per Student Budget":school_summary_data["Per Student Budget"],
                                                "Average Math Score":round(school_summary_data["math_score"],2),
                                                "Average Reading Score":round(school_summary_data["reading_score"],2),
                                                "% Passing Math":school_summary_data["% Passing Math"],
                                                "% Passing Reading":school_summary_data["% Passing Reading"],
                                                "% Overall Passing Rate":school_summary_data["% Overall Passing Rate"]})

#display result
df_school_summary_data
```
